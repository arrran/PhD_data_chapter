```python
"""
From Jenkins 2011 https://doi.org/10.1175/JPO-D-11-03.1 
apply to our channel

From Stewart:
   % Constants

L = 3.35e5;

ci = 2.009e3;

c = 3.974e3;

StantonTS = 2.7e-4;

% Environment

Tf = -2.5;

Ti = -15;

Ti_ef = Tf - L/c -(ci/c)*(Tf-Ti)

Mo = StantonTS/(Tf-Ti_ef)

"""
import numpy as np
import matplotlib.pyplot as plt
```

```python
class plume:
    
    def __init__(self):

        # from table 1
        self.E_0 = 3.6e-2 
        self.C_d = 2.5e-3 
        
        #From stewart
        self.L = 3.35e5
        self.ci = 2.009e3
        self.c = 3.974e3
        
        #diff for different
        self.C_halfd_Gamma_TS = 2.7e-4 #stanton number
    
    def A_0U_func(self):
        """
        A_0U_func = A_0U_func(E_0,C_d,sin_alpha)

        """
        self.A_0U =  (self.sin_alpha / (self.E_0*self.sin_alpha + self.C_d))**(1/3)
        return  self.A_0U
    #XX i dont understand why melt rate factor diff for diff places
    # is f(T_efi, Tif,Gammaa_TS)

    def T_ef_i_func(self):
        """
        T_ef_i_func = T_ef_i_func( T_f, T_i)

        """
        self.T_ef_i =  self.T_f - self.L/self.c -(self.ci/self.c)*(self.T_f-self.T_i)

    def M_0_func(self):
        """
        M_0_func = M_0_func( T_f, T_ef_i, C_halfd_Gamma_TS)

        """
        #self.T_ef_i_func()
        self.M_0 =  self.C_halfd_Gamma_TS / (self.T_f - self.T_ef_i)
        return self.M_0

    def A_0T_func(self):
        """
        A_0T_func = A_0T_func( T_if,sin_alpha, T_ef_i= T_ef_i_func( T_f, T_i),  M_0= M_0_func( T_f, T_ef_i, C_halfd_Gamma_TS))
        
        """
        
        self.A_0T = 1/( 1 - self.M_0*(self.T_ef_i - self.T_if) / (self.E_0*self.sin_alpha) )
        
        return self.A_0T
        
        
    def melt_X0(self):
        # self.A_0U_func()
        # self.M_0_func()
        # self.A_0T_func()
        self.melt = self.A_0U*self.A_0T*self.M_0*self.Q**(1/3)*self.g_DELTA_rhoi_1_3*self.T_a_T_af_X0*3.154e7 #3.154e7 on the end is to convert to m/a
        
        return self.melt
    
    
    #work out T_ef_i to work out T_if for use in finding A_T0 for kamb
    #reverse engineer some vars:
    #M_0,T_f,C_halfd_Gamma_TS >>> T_ef_i
    # A_0T,sin_alpha,M_0,T_ef_i >>> T_if
    # 
    
    def reverse_T_ef_i(self):
        """
        reverse engineer some variables from what jenkins gives
        reverse_T_ef_i = reverse_T_ef_i( M_0,T_f,C_halfd_Gamma_TS)
        
        from: M_0_func = M_0_func( T_f, T_ef_i, C_halfd_Gamma_TS)

        """
        self.T_ef_i = self.T_f - self.C_halfd_Gamma_TS/self.M_0
        return self.T_ef_i
    
    def reverse_T_if(self):
        """
        reverse engineer some variables from what jenkins gives
        reverse_T_if = reverse_T_if(A_0T,sin_alpha,M_0,T_ef_i)
        
        from: A_0T_func = A_0T_func( T_if,sin_alpha, T_ef_i= T_ef_i_func( T_f, T_i),  M_0= M_0_func( T_f, T_ef_i, C_halfd_Gamma_TS))

        """
        
        self.T_if = (1 - self.A_0T)*self.E_0*self.sin_alpha / (self.A_0T*self.M_0) + self.T_ef_i
        return self.T_if
    
    def approx_melt(self,X): 
        self.melt_x = ( (1 + 0.2*(X / self.L_dash))*self.A_0U*self.A_0T*self.M_0*self.Q**(1/3)*self.g_DELTA_rhoi_1_3*self.T_a_T_af_X0) * 3.154e7
        
        return self.melt_x 
    
    def L_dash_func(self):
        self.g_DELTA_rhoi = 0.645**3 #from table 4
        self.g_DELTA_rhoi_ef = 0.002144
        self.L_dash = 5*(self.Q*self.g_DELTA_rhoi)**(2/3)/(self.M_0*self.g_DELTA_rhoi_ef*9.81**(2/3)*self.A_0U*self.A_0T*self.T_a_T_af_X0)
        return  self.L_dash
    
    
```
