```python
bed_gdf = gpd.read_file('/Volumes/arc_04/FIELD_DATA/K8621920/RES/PROCESSED_LINES/picked_bed.shp').query('camp == "KIS2"  & line_name != "lineback2camp"')
# bed_gdf = bed_gdf[bed_gdf.camp =='KIS2']
bed_gdf.keys()
```

```python
plt.figure(figsize=(18.5,15))

plt.scatter(bed_gdf.x.iloc[::-1], bed_gdf.y.iloc[::-1], c=bed_gdf.ice_thickn.iloc[::-1],vmin=800,vmax=400,cmap='gist_rainbow')
cb2 = plt.colorbar()
cb2.set_label('ice base elevation, m.a.s.l.')
plt.xlabel('x, UTM')
plt.ylabel('y, UTM')
# basecolour_surfcontour

plt.xlim([-382000,-380000])
plt.ylim([-726000,-724000])
plt.grid()
# plt.savefig('/Users/home/whitefar/ASSIGNMENTS-PUBLICATIONS-TALKS/channel_data_writeup/PLOTS/radarlines_surfcolour.png')
plt.show()
```

```python
plt.figure(figsize=(18.5,15))
for linename in ['lineupchan', 'linedownchan', 'lineAPRESdown', 'line7p5']:
    plt.plot(bed_gdf[bed_gdf.line_name == linename].x,bed_gdf[bed_gdf.line_name == linename].bed,label=linename)
plt.legend()

plt.xlim([-382000,-380000])
# plt.ylim([-726000,-724000])
plt.grid()
# plt.savefig('/Users/home/whitefar/ASSIGNMENTS-PUBLICATIONS-TALKS/channel_data_writeup/PLOTS/radarlines_surfcolour.png')
plt.show()
```
