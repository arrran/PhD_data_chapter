```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d, RectBivariateSpline, griddata
```

```python
path = '/Users/home/whitefar/DATA/FIELD_ANT_19/POST_FIELD/RADAR_MAP/PRACTICE_INTERP/'
points = gpd.read_file(path+'buck_allpoints.shp',crs=2193).drop(['Latitude (','Longitude'],axis=1).rename(columns={'Height (Gl': 'height'})
# points.to_crs(2193)
```

```python
def scatter_df(df):
    fig = plt.figure(figsize=(20,8))
    plt.scatter(df.geometry.x, df.geometry.y, c=df.height )
    plt.clim([df.height.min(),df.height.max()])
    plt.colorbar(label='height')
    plt.xlim(df.geometry.x.min(),df.geometry.x.max())
    plt.ylim(df.geometry.y.min(),df.geometry.y.max())
    plt.show()
def scatter_xyz(x,y,z):
    fig = plt.figure(figsize=(20,8))
    plt.scatter(x, y, c=z )
    plt.clim([z.min(),z.max()])
    plt.colorbar(label='height')
    plt.xlim(x.min(),x.max())
    plt.ylim(y.min(),y.max())
    plt.show()
```

## Spatial Interpolation of scattered data

Once, the data has been filtered we can interpolate the points data to a regular grid to allow for different kinds of analysis and mass-balance computations. Here, we will present three types of different interpolation schemes and discuss pro's and cons of each approach

Interpolation algorithms:

- Median
- Gaussian Weighted Average
- Least-Squares Collocation (Kriging)


```python
coords = np.hstack([gdf.geometry.x.to_numpy().reshape(210,1), gdf.geometry.y.to_numpy().reshape(210,1)])
```

```python
xx = np.arange(points.geometry.x.min(), points.geometry.x.max(), 1)
yy = np.arange(points.geometry.y.min(), points.geometry.y.max(), 1)
Xi,Yi = np.meshgrid(xx,yy)
```

```python
z_out = griddata(coords, gdf.height.to_numpy(),(Xi,Yi))
```

```python
extent = [points.geometry.x.min(), points.geometry.x.max(),points.geometry.y.min(), points.geometry.y.max()]
plt.imshow(z_out,extent =extent)
plt.colorbar(label='height')
plt.show()
```
