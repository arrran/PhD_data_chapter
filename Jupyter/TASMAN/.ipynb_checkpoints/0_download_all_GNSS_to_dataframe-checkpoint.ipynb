# Download GNSS data for May 2016

This notepad uses my function GNSS_DATA, load_GNSSdata_for_site to reorganise data from GNSS-output format spread across heaps of files to a nice pandas dataframe.

it gets data from (2016, 5, 1) to (2016, 5, 18) as in https://doi.org/10.1017/aog.2019.20 sam taylor-offords seismic paper



```python
import os
import sys
import datetime

sys.path.append(os.path.abspath('/Users/home/whitefar/DATA/code/'))
```

```python
from GNSS_DATA import load_GNSSdata_for_site
```

```python
start = datetime.datetime(2016, 5, 1).timestamp()
finish = datetime.datetime(2016, 5, 18).timestamp()
```

```python
site_units = [["tal1","arc2"],   #arc*s are the GNSS units, ta*s are the sites
             ["tal2","arc1"],
             ["tac1","arc1"],
             ["tac2","arc5"],
             ["tac3","arc4"],
             ["tar1","arc8"],
             ["tar2","arc6"]
             ]

sites_with_data = []

for site, unit in site_units:
    print(site,unit)
    site_df, has_data = load_GNSSdata_for_site([site,unit],start,finish)
    
    if has_data == True:
        site_df.sort_values(by=['Timestamp'], inplace=True) #puts them in time order, 
        site_df.reset_index(drop=True, inplace=True)    #resets the index after putting in time order

        site_df.to_csv(f"/Volumes/arc_04/whitefar/DATA/TASMAN/GNSS_ABSOLUTE/{site}{unit}_may1-18_2016.csv")
        
        sites_with_data.append([site,unit])
```

```python
print(sites_with_data)
```
