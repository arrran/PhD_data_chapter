### Above is a list of kmls in the temp folder
we can now use the same function as 9_cryosat to just save the intersection

```python
area_wanted_path = "/Users/home/whitefar/DATA/FIELD_ANT_19/study_area_buffer.shp"
area_wanted_gdf = gpd.read_file(area_wanted_path)
```

```python
def kml2shp_intersection(area_wanted_gdf, input_kml, output_folder):
    """
    This reads a particular kml file, with heaps of lines
    then converts them to epsg3031 then gets intersection of the lines with an area
    then outputs the intersection
    """
    
    area_wanted_polygon = area_wanted_gdf.iloc[0].geometry # THIS IS VERY IMPORTANT
    # CANNOT BE A SERIES OR EACH ROW ENTRY COMPARES TO EACH ROW ENTRY
    
    path, fnameext = os.path.split(input_kml)
    fname, ext = os.path.splitext(fnameext)
    
    path_gdf = gpd.read_file(input_kml, driver='LIBKML').to_crs({'init' :'epsg:3031'}) #read path and convert datum
    
    path_gdf["intersection"] = path_gdf.intersection(area_wanted_polygon)   #make a new column of the intersection
    path_gdf = path_gdf.set_geometry('intersection')  # set this as the main "geometry"
    path_gdf = path_gdf.drop('geometry', axis=1)   #delete the original line
    path_gdf.to_file(os.path.join(output_folder,"CRYOSAT2_"+ fname + ".shp"))
    print("written a line to ",os.path.join(output_folder, "CRYOSAT2_" + fname + ".shp"))
            
```

```python
for i, kml in enumerate(kml_paths):
    
    kml2shp_intersection(area_wanted_gdf, kml, output_folder)
    
    print(i+1,"/",len(kml_paths))
```
