### Workflow

1. read kml:

gpd.read_file()

2. convert datum

.to_crs({'init' :'epsg:3031'})

3. does the line intersect field buffer

4. save interseection as shp file


```python
import numpy as np
import zipfile

import os
import sys
import geopandas as gpd
import shapely as sp
import matplotlib.pyplot as plt
import fiona
import pyproj


fiona.supported_drivers['libkml'] = 'r'
fiona.supported_drivers['LIBKML'] = 'r'

#working_dir = '/Volumes/arc_02/REMOTE_SENSING/'
directory = "/Volumes/arc_02/REMOTE_SENSING/CRYOSAT2/GROUND_TRACKS/SHAPEFILES"

sys.path.append(os.path.abspath(directory))
sys.path.append(os.path.abspath('/Users/home/whitefar/DATA/code/'))
os.chdir(directory)

```

```python
output_folder = "/Volumes/arc_02/REMOTE_SENSING/CRYOSAT2/GROUND_TRACKS/SHAPEFILES/FIELD_SITE_ANT_19"
input_file = "/Volumes/arc_02/REMOTE_SENSING/CRYOSAT2/GROUND_TRACKS/ORIGINALS/CRYOSAT2_ORBIT_SWATH_20190711_040000_20200713_110000.KML"

area_wanted_path = "/Users/home/whitefar/DATA/FIELD_ANT_19/study_area_buffer.shp"
area_wanted_gdf = gpd.read_file(area_wanted_path)
```

```python
os.path.isfile(input_file)
```

```python
# get list of all the paths within the kml file
cryosat_pathnames = fiona.listlayers(input_file)[1:]  #first in list is the header
```

```python
def export_only_intersection(area_wanted_gdf, path_name, input_kml, output_folder):
    """
    This reads a particular path from a sublayer in a kml file, 
    then converts it to epsg3031 then checks to see if if overlaps an area
    then outputs the intersection if it does
    """
    
    path_gpd = gpd.read_file(input_file, driver='LIBKML', layer=path_name).to_crs({'init' :'epsg:3031'}) #read path and convert datum
    
    if bool(path_gpd.intersects(area_wanted_gdf)[0]) is True:
        path_gpd.intersection(area_wanted_gdf).to_file(os.path.join(output_folder,"CRYOSAT2_"+ path_name + ".shp"))
        outputted = 1
        print("written a line to ",os.path.join(output_folder, "CRYOSAT2_" + path_name + ".shp"))
    else:
        outputted = 0
        
    return outputted
    
```

```python
num_lines = 0

for i, path_name in enumerate(cryosat_pathnames):
    
    out = export_only_intersection(area_wanted_gdf, path_name, input_file, output_folder)
    
    num_lines = num_lines + out
    
    print(i+1,"/",len(cryosat_pathnames))
    
print("number of lines written which intersect area is " + num_lines)
    
```
