```python
area_wanted_path = "/Users/home/whitefar/DATA/FIELD_ANT_19/study_area_buffer.shp"
#area_wanted_path = "/Users/home/whitefar/DATA/FIELD_ANT_19/test/antarcticaish.shp"

output_folder = "/Volumes/arc_02/REMOTE_SENSING/ICESAT2/REFERENCE_GROUND_TRACKS/SHAPEFILES/cut_ANT19_GROUNDTRACKS"

area_wanted_gdf = gpd.read_file(area_wanted_path)
```

```python
def export_only_intersection(area_wanted_gdf, shp_path, output_folder):
    """starts with paths, outputs  shapefile lines"""
    
    fpath, fname = os.path.split(shp_path)
    
    line = gpd.read_file(shp_path).to_crs({'init' :'epsg:3031'})
    
    if bool(line.intersects(area_wanted_gdf)[0]) is True:
        line.intersection(area_wanted_gdf).to_file(os.path.join(output_folder, fname))
        outputted = 1
        print("written a line to ",os.path.join(output_folder, fname))
    else:
        outputted = 0
        

    return outputted
```

## Cut to the intersection over all of the shapefiles

```python
num_lines = 0

for i, shp_path in enumerate(shp_paths):
    
    out = export_only_intersection(area_wanted_gdf, shp_path, output_folder)
    
    num_lines = num_lines + out
    
    print(i,"/",len(shp_paths))
    
print(num_lines)
    
```

```python
#list files in  a directory with shp

file = os.listdir(output_folder)
path = [os.path.join(output_folder, i) for i in file if ".shp" in i]
len(path)
```
